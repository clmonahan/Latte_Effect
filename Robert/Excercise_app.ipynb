```python
import pandas as pd
import numpy as np
```

```python
data = "excercise3.csv"
data_df = pd.read_csv(data)
data_df.head()
```

```python
data_df.info()
```

```python
df = data_df.astype({"Excercise":str, "100":int , "125":int , "150":int , "175":int , "200":int , "250":int , "300":int})
```

```python
# convert just columns "a" and "b"
# df[["a", "b"]] = df[["a", "b"]].apply(pd.to_numeric)

data_df[["100", "125", "150", "175", "200", "250", "300"]] = data_df[["100", "125", "150", "175", "200", "250", "300"]].apply(pd.to_numeric, errors='ignore')
```

```python
data_df.info()
```

```python
numeric_df = pd.DataFrame()
```

```python
cols = data_df.columns
for col in cols:
    if col == 'Exercise':
        numeric_df[col] = data_df[col]
    else:
        numeric_df[col] = data_df[col].map(lambda row: pd.to_numeric(row, errors='coerce'))
```

```python
numeric_df.info()
```

```python
numeric_df.dropna(inplace=True)
```

```python
numeric_df.info()
```

```python
numeric_df.head()
```

```python
for col in cols:
    if col == 'Exercise':
        continue
    numeric_df[col] = 500 / numeric_df[col]
```
