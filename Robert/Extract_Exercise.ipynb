Extract CSVs into DataFrames

```python
source_file = os.path.join(".","calperminute.csv")
cal_df = pd.read_csv(source_file)
cal_df.head()
```

```python
new_cal_df = cal_df[['Exercise', '100', '125', '150', '175', '200', '250', '300']]
new_cal_df.head()
```

```python
new_cal_df.to_csv(r'Output\calperminute.csv')
```
