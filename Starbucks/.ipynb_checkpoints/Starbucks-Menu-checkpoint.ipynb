```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
from sqlalchemy import event, create_engine
from secrets import mysqlpassword
```

```python
#Read in Files 
drinks_menu = pd.read_csv("Resources/starbucks-menu-nutrition-drinks.csv")
expanded_drink_menu = pd.read_csv("Resources/starbucks_drinkMenu_expanded.csv")
food_menu = pd.read_csv("Resources/starbucks-menu-nutrition-food.csv")
```

```python
#Remove Rows Without Calorie Column Data and Rename Column
drinks_menu = drinks_menu[drinks_menu.Calories != "-"]
drinks_menu =drinks_menu.rename(columns = {'Unnamed: 0':'Item'})
drinks_menu= drinks_menu[['Item', 'Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein']]

drinks_menu.rename(columns={'Protein':'Protein (g)'}, inplace=True)


drinks_menu.head()
```

```python
food_menu.head()
```

```python
list(food_menu)
```

```python
food_menu.rename(columns={ ' Fat (g)':'Fat (g)',
                          ' Carb. (g)':'Carb. (g)',
                          ' Fiber (g)':'Fiber (g)',
                               ' Protein (g)': 'Protein (g)',
                         ' Calories':'Calories'}, 
                 inplace=True)
```

```python
expanded_drink_menu['Item'] = expanded_drink_menu['Beverage'] + '-' + expanded_drink_menu['Beverage_prep']
```

```python
new_drink_menu = expanded_drink_menu[['Item','Calories',
                                      ' Total Fat (g)', ' Total Carbohydrates (g) ',
                                      ' Dietary Fibre (g)', ' Protein (g) ']]
new_drink_menu.head()
```

```python
new_drink_menu.rename(columns={' Total Fat (g)':'Fat (g)',
                          ' Total Carbohydrates (g) ':'Carb. (g)',
                          ' Dietary Fibre (g)':'Fiber (g)',
                               ' Protein (g) ': 'Protein (g)'}, 
                 inplace=True)
```

```python
new_drink_menu.head()
```

```python
drinks_menu.head()
```

```python
new_drink_menu= drinks_menu[['Item', 'Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)']]
new_drink_menu.head()
```

```python
merged_drinks = new_drink_menu.append(drinks_menu)
list(merged_drinks)
```

```python
full_menu = merged_drinks.append(food_menu)
```
